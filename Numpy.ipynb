NumPy (short for Numerical Python) is a Python library used for performing efficient numerical computations. It provides tools to work with large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them.



*   What is a Library?
*   *List item*





Key Features of NumPy
*   N-dimensional Arrays: ndarray -> multidimensional arrays
*   ndarray -> speed and memory [built on C] as compared to List


*   Mathematical Operations:mean, sum, max, min, standard deviation, etc.
*   Allows vectorized operations -> NO discuss while I implement







```python
import numpy as np #aliasing
```

```python
temp1 = np.array([22, 34,45,25,36]) #ndarray
#temp = [22, 34,56,78,90]
#temp1
#print(temp1)
print("Temperatures:", temp1)
```

```python
max_temp = np.max(temp1)
print("Hottest Day Temperature: ",max_temp)
```

```python
min_temp = np.min(temp1)
print("Coldest Day Temperature: ",min_temp)
```

```python
average_temp = np.mean(temp1)
print("Average Day Temperature: ",average_temp)
```

```python
# Identify Days with Temp > Average Temperature
day_above_avg_temp = temp1 > average_temp
print("Days with temp greater than Average: ",day_above_avg_temp)
```
